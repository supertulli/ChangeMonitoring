```python
# Distance matrix

from tqdm import tqdm
import numpy as np

from  mpl_toolkits.mplot3d import Axes3D

from scipy.stats import entropy
from scipy.special import rel_entr, kl_div
```

```python
dist_matrix_size = rel_freq_table.shape[0]
dist_matrix = np.zeros((dist_matrix_size,dist_matrix_size), dtype='float')
```

```python

def KL_div(p,q):
    return np.dot(p,np.log2(p/q))
```

```python
from multiprocessing.pool import Pool

def get_JSD_distance(i, j, source_df):
    P_i, P_j = source_df.iloc[i].to_numpy(), source_df.iloc[j].to_numpy()
    M = (P_i + P_j)/2
    return i, j , np.sqrt((entropy(P_i,M, base=2)+entropy(P_j,M, base=2))/2)
    # dist_matrix[i][j] = np.sqrt((entropy(P_i,M, base=2)+entropy(P_j,M, base=2))/2)
    # dist_matrix[j][i] = dist_matrix[i][j]

def get_dist_matrix(source_df):
    
    dist_matrix_size = source_df.shape[0]
    dist_matrix = np.zeros((dist_matrix_size,dist_matrix_size), dtype=source_df.dtypes.mode()[0].name.lower())
    
    with Pool() as pool:
        inputs = [(i, j, source_df) for i in range(dist_matrix_size) for j in range(i+1,dist_matrix_size)]
        for i, j, dist in pool.starmap(get_JSD_distance, tqdm(inputs)):
            dist_matrix[i][j] = dist
            dist_matrix[j][i] = dist_matrix[i][j]

        
    
    return dist_matrix
```

```python
age_dist_matrix = get_dist_matrix(rel_freq_table)
```

```python
rel_freq_table.dtypes.mode()[0].name
```

```python

for i in tqdm(range(dist_matrix_size)):
    # print(i)
    for j in range(i+1,dist_matrix_size):
        P_i, P_j = rel_freq_table.iloc[i].to_numpy(), rel_freq_table.iloc[j].to_numpy()
        M = (P_i + P_j)/2
        # print('i:', i, '  j:', j)
        dist_matrix[i][j] = np.sqrt((entropy(P_i,M, base=2)+entropy(P_j,M, base=2))/2)
        dist_matrix[j][i] = dist_matrix[i][j]
```
